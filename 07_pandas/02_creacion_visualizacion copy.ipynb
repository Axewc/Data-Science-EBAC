¿Qué es un archivo .CSV? 
- CSV = “Comma-Separated Values” (Archivo de valores separados por coma) 
- Diseñado para un manejo óptimo en DataFrames 
- La mayoría de los programas de bases de datos y hoja de cálculo puedan usarlos y/o crearlos. 
- Debido a la flexibilidad de su formato pueden guardar grandes bases de datos en poco espacio en disco.

### Visualización de datos .CSV 

__Ejercicio:__ Genere un histograma de frecuencias para la variable “sepal.length” importadas del archivo “Iris.csv”. A continuación genera un gráfico de barras que muestra la longitud promedio del pétalo por cada tipo de flor.



```python
# Creación de un histograma
iris["sepal.length"].hist()
```

```python
subintervalos = int(round(len(iris) ** 0.5,0)) # Regla de Sturges
subintervalos
```

```python
iris["sepal.length"].hist(bins=subintervalos)
```

```python
iris_tipo = iris.groupby("variety") ["petal.length"].mean()
iris_tipo
```

```python
iris_tipo.plot(kind='bar')
```
